## Теоретическая часть: метод синтетического контроля

### Почему используется Synthetic Control Method (SCM)

Для оценки эффекта строительства тюрем в Техасе используется метод синтетического контроля (Synthetic Control Method, SCM). Данный метод применяется в ситуациях, когда воздействие получает одна единица наблюдения — в данном случае штат Техас, — и необходимо оценить, каким был бы исход без вмешательства.

Идея SCM состоит в построении контрфактического сценария с помощью взвешенной комбинации других штатов (donor pool), которая максимально точно воспроизводит характеристики Техаса в доинтервенционный период.

Формально causal-эффект можно записать как:

$$
\alpha_{1t} = Y_{1t}^{obs} - Y_{1t}^{N}
$$

где:
- $Y_{1t}^{obs}$ — наблюдаемый уровень incarceration,
- $Y_{1t}^{N}$ — контрфактический уровень без вмешательства.

Так как $Y_{1t}^{N}$ не наблюдается, он аппроксимируется через синтетический контроль:

$$
\hat{Y}_{1t}^{N} = \sum_{j=2}^{J+1} w_j Y_{jt}
$$

где веса удовлетворяют ограничениям:

$$
w_j \geq 0, \quad \sum_{j=2}^{J+1} w_j = 1
$$

Таким образом, синтетический Техас — это выпуклая комбинация других штатов.

---

### Почему SCM подходит для данной задачи

SCM особенно подходит для данного кейса по следующим причинам:

1. Воздействие применяется к одной единице (Техас).
2. Нет очевидного «идеального» контрольного штата.
3. Можно точно воспроизвести динамику до вмешательства.

В отличие от Difference-in-Differences, SCM не требует предположения о параллельных трендах, а вместо этого напрямую подбирает веса, минимизируя различия в pre-treatment периоде.

---

### Ключевые предположения

#### 1. Отсутствие интерференции (SUTVA)

Предполагается, что строительство тюрем в Техасе не влияет на другие штаты.

#### 2. Достаточный доинтервенционный период

Период 1985–1992 используется для подбора весов. Важно, чтобы синтетический контроль хорошо совпадал с Техасом в этот период.

#### 3. Отсутствие других шоков

После 1993 года не должно быть других уникальных факторов, влияющих на incarceration в Техасе.

#### 4. Представимость через донорный пул

Техас должен быть аппроксимируем как комбинация других штатов.

---

### Выбор предикторов

В модели используются следующие переменные:

- $poverty$ — уровень бедности  
- $income$ — доход на душу населения  
- $black$ — доля чернокожего населения  
- $perc1519$ — доля молодежи  
- $alcohol$ — потребление алкоголя  
- $aidscapita$ — случаи СПИДа  

Также используются лаги зависимой переменной $bmprison$.

Обоснование:

- Экономические переменные ($poverty$, $income$) влияют на преступность  
- Демография ($black$) напрямую связана с изучаемой группой  
- Возрастная структура ($perc1519$) влияет на криминальную активность  
- Социальные факторы ($alcohol$, $aidscapita$) отражают поведенческий контекст  

---

### Почему 1993 год — точка разрыва

1993 год выбран как момент вмешательства, поскольку именно тогда при губернаторе Энн Ричардс началась масштабная программа строительства тюрем.

Это обеспечивает:

- четкое разделение на pre-treatment (1985–1992)  
- и post-treatment (1993–2000)  

Таким образом, 1993 год является естественной точкой начала воздействия.

---

### Интерпретация эффекта

Оценка эффекта производится как разница между фактическим и синтетическим Техасом:

$$
\hat{\alpha}_{1t} = Y_{Texas,t} - Y_{Synthetic\ Texas,t}, \quad t \geq 1993
$$

Если после 1993 года фактическая траектория выше синтетической, это означает положительный эффект политики на уровень incarceration.

---

### Вывод

Метод синтетического контроля позволяет построить правдоподобный контрфактический сценарий для Техаса. При выполнении ключевых предположений разница между фактическим и синтетическим Техасом после 1993 года интерпретируется как causal-эффект строительства тюрем.
